```python
## TEMPERATURE ESTIMATION from OES
## BASED ON N2(CB) TRANSITION
## Author: Dogan Gidon, 2018/07/12

#import tools
import numpy as np
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
#from scipy import io as sio
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import Lasso

################################################################################
# LOAD TRAINING AND TEST DATA
# first 51 columns are the OES signal
# last two columns are rotational temeprature and vibrational temperature

df_train=pd.read_csv('dat_train.csv') ##import training data
df_test=pd.read_csv('dat_test.csv') ##import test data
df_test_d=pd.read_csv('dat_test_dsep.csv') ##import test data for changing seperation dsitance
df_test_o=pd.read_csv('dat_test_o2.csv') ##import test data for changing O2 concentration


```

```python
## visualize independent variabiles i.e. OES peaks

import matplotlib.pyplot as plt
l=np.arange(0,len(df_train))
l2=np.arange(0,len(df_test))
plt.figure()
plt.plot(np.transpose(np.array(df_train.iloc[:,0:51])))

plt.show()
```

```python

#select training and test input data
Xtest=np.array(df_test.iloc[:,0:51])
Xtrain=np.array(df_train.iloc[:,0:51])

Xtest_p=np.array(df_test.iloc[:,0:52])
Xtrain_p=np.array(df_train.iloc[:,0:52])

#select training and test output data
# selecting Trot and Tvib and operating conditions 
Ytest=np.array(df_test.iloc[:,53:55])
Ytrain=np.array(df_train.iloc[:,53:55])   

Ytest=np.array(df_test.iloc[:,51:56])
Ytrain=np.array(df_train.iloc[:,51:56])    

#normalize
Ytrain[:,2]=Ytrain[:,2]/550.
Ytrain[:,3]=Ytrain[:,3]/4000.

Ytest[:,2]=Ytest[:,2]/550.
Ytest[:,3]=Ytest[:,3]/4000.

## additional test data with changing seperation distance
Xtest_p_d=np.array(df_test_d.iloc[:,0:52])
Xtest_d=np.array(df_test_d.iloc[:,0:51])
Ytest_d=np.array(df_test_d.iloc[:,53:55])

Ytest_d[:,0]=Ytest_d[:,0]/550.
Ytest_d[:,1]=Ytest_d[:,1]/4000.

## additional test data with changing additive O2 concentraiton
Xtest_p_o=np.array(df_test_o.iloc[:,0:52])
Xtest_o=np.array(df_test_o.iloc[:,0:51])
Ytest_o=np.array(df_test_o.iloc[:,53:55])

Ytest_o[:,0]=Ytest_o[:,0]/550.
Ytest_o[:,1]=Ytest_o[:,1]/4000.

#select the operating conditions in dsep and O1 concentration
dsep=np.array(df_test_d.iloc[:,57])
o_conc=np.array(df_test_o.iloc[:,56])

df_test.shape
plt.figure()
plt.plot(Ytest)
plt.show()
```

```python
#split the data training and test sections with 20% reserved for testing
from sklearn.model_selection import train_test_split

X_train, X_test, Y_trains, Y_tests = train_test_split(Xtrain, Ytrain, test_size=0.2, random_state=0)
```

```python
Y_train=Ytrain[:,2:4]
Y_test=Ytest[:,2:4]

X_train=Xtrain
X_test=Xtest
```

```python
#plot training data
#Trot and Tvib
plt.figure()
plt.plot(Y_train[:,0]*550,label='Trot')
plt.plot(Y_train[:,1]*4000,label='Tvib')
plt.legend()
#Operating conditions
plt.figure()
plt.plot(Ytrain[:,0],label='Power')
plt.plot(Ytrain[:,1],label='Flow')
plt.plot(Ytrain[:,4],label='Substrate type') #0 is glass 1 is metal
plt.legend()
plt.show()
```

```python
#compute high order polynomial bases up to order 9
Xtrain2=np.hstack((X_train,X_train**2))
Xtrain3=np.hstack((X_train,X_train**2,X_train**3))
Xtrain4=np.hstack((X_train,X_train**2,X_train**3,X_train**4))
Xtrain5=np.hstack((X_train,X_train**2,X_train**3,X_train**4,X_train**5))
Xtrain6=np.hstack((X_train,X_train**2,X_train**3,X_train**4,X_train**5,X_train**6))
Xtrain7=np.hstack((X_train,X_train**2,X_train**3,X_train**4,X_train**5,X_train**6,X_train**7))
Xtrain8=np.hstack((X_train,X_train**2,X_train**3,X_train**4,X_train**5,X_train**6,X_train**7,X_train**8))
Xtrain9=np.hstack((X_train,X_train**2,X_train**3,X_train**4,X_train**5,X_train**6,X_train**7,X_train**8,X_train**9))


Xtest2=np.hstack((X_test,X_test**2))
Xtest3=np.hstack((X_test,X_test**2,X_test**3))
Xtest4=np.hstack((X_test,X_test**2,X_test**3,X_test**4))
Xtest5=np.hstack((X_test,X_test**2,X_test**3,X_test**4,X_test**5))
Xtest6=np.hstack((X_test,X_test**2,X_test**3,X_test**4,X_test**5,X_test**6))
Xtest7=np.hstack((X_test,X_test**2,X_test**3,X_test**4,X_test**5,X_test**6,X_test**7))
Xtest8=np.hstack((X_test,X_test**2,X_test**3,X_test**4,X_test**5,X_test**6,X_test**7,X_test**8))
Xtest9=np.hstack((X_test,X_test**2,X_test**3,X_test**4,X_test**5,X_test**6,X_test**7,X_test**8,X_test**9))
```

```python
#run 10-fold cross-valdiation across different basis orders and alpha values
#for the Lasso algorithm
cv2 = KFold(n_splits=10, random_state=1, shuffle=True)
r=np.linspace(3e-6,3e-5,10)

sc1=[]
std1=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, X_train, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc1 +=[np.mean(scores)]
    std1 +=[np.std(scores)]
print(' ')

sc2=[]
std2=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain2, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc2 +=[np.mean(scores)]
    std2 +=[np.std(scores)]
print(' ')
sc3=[]
std3=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain3, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc3 +=[np.mean(scores)]
    std3 +=[np.std(scores)]    
print(' ')
sc3=[]
std3=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain3, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc3 +=[np.mean(scores)]
    std3 +=[np.std(scores)]      
print(' ')
sc4=[]
std4=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain4, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc4 +=[np.mean(scores)]
    std4 +=[np.std(scores)]      
print(' ')
sc5=[]
std5=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain5, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc5 +=[np.mean(scores)]
    std5 +=[np.std(scores)]      
print(' ')
sc6=[]
std6=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain6, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc6 +=[np.mean(scores)]
    std6 +=[np.std(scores)]                      
print(' ')
sc7=[]
std7=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain7, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc7 +=[np.mean(scores)]
    std7 +=[np.std(scores)]      
print(' ')
sc8=[]
std8=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain8, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc8 +=[np.mean(scores)]
    std8 +=[np.std(scores)]     
print(' ')
sc9=[]
std9=[]
for aa in r:   
    las_cov2=Lasso(alpha=aa)
    scores = cross_val_score(las_cov2, Xtrain9, Y_train,cv=cv2)
    print(aa,np.mean(scores),'pm',np.std(scores))
    sc9 +=[np.mean(scores)]
    std9 +=[np.std(scores)] 
```

```python
#plot cross-validation results

plt.figure()
plt.plot(r,sc1,label='1')
plt.plot(r,sc2,label='2')
plt.plot(r,sc3,label='3')
plt.plot(r,sc4,label='4')
plt.plot(r,sc5,label='5')
plt.plot(r,sc6,label='6')
plt.plot(r,sc7,label='7') 
plt.plot(r,sc8,label='8')
plt.plot(r,sc9,label='9')
plt.xlabel('Alpha')
plt.ylabel('Validation score (R^2)')
plt.legend()
plt.show()
```

```python
#pick the lowest order basis which gives good validation score
las_cov=Lasso(alpha=7.1e-6)
las_cov.fit(Xtrain4,Y_train)
Yp=las_cov.predict(Xtest4)
```

```python
#Compare tested and predicted Trot and Tvib
from sklearn.metrics import r2_score, mean_squared_error
plt.figure()
plt.plot(Yp[:,0]*550,'o',label='Trot Predicted')
plt.plot(Y_test[:,0]*550,label='Trot Fitted from Spec')

plt.plot(Yp[:,1]*4000,label='Tvib Predicted')
plt.plot(Y_test[:,1]*4000,label='Tvib Fitted from Spec')
plt.legend()
plt.ylabel('T(K)')
plt.xlabel('Measurement Instance')
plt.show()
```

```python
# create appropriate order test data for data
# with changing seperation distance and changing O2 conc

Xtest4_d=np.hstack((Xtest_d,Xtest_d**2,Xtest_d**3,Xtest_d**4))
Xtest4_o=np.hstack((Xtest_o,Xtest_o**2,Xtest_o**3,Xtest_o**4))

# predit Trot and Tvib for the above test data
Ypd=las_cov.predict(Xtest4_d)
Ypo=las_cov.predict(Xtest4_o)
```
